```python
import pandas as pd
import matplotlib.pyplot as plt

import matplotlib 
matplotlib.rc('xtick', labelsize=22) 
matplotlib.rc('ytick', labelsize=22)

# Load CSV
filename = "PID_motor_data_20250405_002719.csv"
df = pd.read_csv(filename)

# Prompt for filenames
plot1_name = input("Enter filename for Speed Tracking plot (without extension): ").strip() + ".pdf"
plot2_name = input("Enter filename for PID Gains plot (without extension): ").strip() + ".pdf"
plot3_name = input("Enter filename for Error Histogram plot (without extension): ").strip() + ".pdf"

# Time and signals
times = df['Time (s)']
desired = df['Desired Speed (RPM)']
measured = df['Measured Speed (RPM)']
error = df['Error (RPM)']

# PID gains
kp = df['Kp']
ki = df['Ki']
kd = df['Kd']

# === Plot 1: Speed Tracking with ±10% Tolerance ===
# ±10% dynamic threshold band around Desired Speed
threshold_lower = desired * 0.90
threshold_upper = desired * 1.10


plt.figure(figsize=(9, 7))
plt.plot(times, desired, label='Desired Speed (RPM)', color='green', lw=3)
plt.plot(times, measured, label='Measured Speed (RPM)', color='blue', lw=3)
plt.plot(times, error, label='Error (RPM)', color='red', linestyle='--', lw=3)
plt.fill_between(times, threshold_lower, threshold_upper, color="gray", alpha=0.3, label="±10% Tolerance Band")


plt.xlabel('Time (s)', fontsize=24)
plt.ylabel('Speed / Error (RPM)', fontsize=24)
#plt.title('Speed Tracking with ±10% Tolerance and Settling Time', fontsize=16)
plt.legend(fontsize=20, loc='upper right')
plt.ylim(-4000, 10000)
plt.grid(True)
plt.xticks(fontsize=22)
plt.yticks(fontsize=22)
plt.tight_layout()
plt.savefig(plot1_name, format="pdf", bbox_inches="tight")
plt.show()
print(f"Speed tracking plot saved as: {plot1_name}")

# === Plot 2: PID Gains Over Time ===
plt.figure(figsize=(9, 7))
plt.plot(times, kp, label='Kp', color='blue', lw=3)
plt.plot(times, ki, label='Ki', color='red', lw=3)
plt.plot(times, kd, label='Kd', color='green', lw=3)
plt.xlabel('Time (s)', fontsize=24)
plt.ylabel('PID Gain Value', fontsize=24)
#plt.title('PID Gains Over Time', fontsize=16)
plt.legend(fontsize=20)
plt.ylim(-10, 70)
plt.grid(True)
plt.xticks(fontsize=22)
plt.yticks(fontsize=22)
plt.tight_layout()
plt.savefig(plot2_name, format="pdf", bbox_inches="tight")
plt.show()
print(f"PID gains plot saved as: {plot2_name}")

# === Plot 3: Histogram of Error ===
plt.figure(figsize=(9, 7))
plt.hist(error, bins=50, color='blue', edgecolor='black')
plt.axvline(0, color='black', linestyle='--')
plt.axvline(0.10 * desired.max(), color='gray', linestyle='--', label='+10%')
plt.axvline(-0.10 * desired.max(), color='gray', linestyle='--', label='-10%')
plt.xlabel('Error (RPM)', fontsize=24)
plt.ylabel('Frequency', fontsize=24)
#plt.title('Histogram of Speed Error', fontsize=16)
plt.legend(fontsize=20)
plt.grid(True)
plt.xticks(fontsize=22, rotation=45)
plt.yticks(fontsize=22)
plt.tight_layout()
plt.savefig(plot3_name, format="pdf", bbox_inches="tight")
plt.show()
print(f"Error histogram saved as: {plot3_name}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load CSV
filename = "motor_speed_data_ANN_20250405_031459.csv"
df = pd.read_csv(filename)

# Prompt for filenames
plot1_name = input("Enter filename for Speed Tracking plot (without extension): ").strip() + ".pdf"
plot2_name = input("Enter filename for PID Gains plot (without extension): ").strip() + ".pdf"
plot3_name = input("Enter filename for Error Histogram plot (without extension): ").strip() + ".pdf"
plot4_name = input("Enter filename for Loss plot (without extension): ").strip() + ".pdf"

# Time and signals
times = df['Time (s)']
desired = df['Desired Speed (RPM)']
measured = df['Measured Speed (RPM)']
error = df['Error (RPM)']
kp = df['Kp']
ki = df['Ki']
kd = df['Kd']
loss = df['Loss']

# === Plot 1: Speed Tracking with ±10% Tolerance ===
threshold_lower = desired * 0.90
threshold_upper = desired * 1.10

plt.figure(figsize=(9, 7))
plt.plot(times, desired, label='Desired Speed (RPM)', color='green', lw=3)
plt.plot(times, measured, label='Measured Speed (RPM)', color='blue', lw=3)
plt.plot(times, error, label='Error (RPM)', color='red', linestyle='--', lw=3)
plt.fill_between(times, threshold_lower, threshold_upper, color="gray", alpha=0.3, label="±10% Tolerance Band")
plt.xlabel('Time (s)', fontsize=24)
plt.ylabel('Speed / Error (RPM)', fontsize=24)
#plt.title('Speed Tracking with ±10% Tolerance', fontsize=16)
plt.legend(fontsize=20, loc='upper right')
plt.ylim(-4000, 10000)
plt.grid(True)
plt.xticks(fontsize=22)
plt.yticks(fontsize=22)
plt.tight_layout()
plt.savefig(plot1_name, format="pdf", bbox_inches="tight")
plt.show()
print(f"Speed tracking plot saved as: {plot1_name}")

# === Plot 2: PID Gains Over Time ===
plt.figure(figsize=(9, 7))
plt.plot(times, kp, label='Kp', color='blue', lw=2)
plt.plot(times, ki, label='Ki', color='red', lw=2)
plt.plot(times, kd, label='Kd', color='green', lw=2)
plt.xlabel('Time (s)', fontsize=24)
plt.ylabel('PID Gain Value', fontsize=24)
#plt.title('PID Gains Over Time', fontsize=16)
plt.legend(fontsize=20)
plt.ylim(-10, 70)
plt.grid(True)
plt.xticks(fontsize=22)
plt.yticks(fontsize=22)
plt.tight_layout()
plt.savefig(plot2_name, format="pdf", bbox_inches="tight")
plt.show()
print(f"PID gains plot saved as: {plot2_name}")

# === Plot 3: Histogram of Error ===
plt.figure(figsize=(9, 7))
plt.hist(error, bins=50, color='blue', edgecolor='black')
plt.axvline(0, color='black', linestyle='--')
plt.axvline(0.10 * desired.max(), color='gray', linestyle='--', label='+10%')
plt.axvline(-0.10 * desired.max(), color='gray', linestyle='--', label='-10%')
plt.xlabel('Error (RPM)', fontsize=24)
plt.ylabel('Frequency', fontsize=24)
#plt.title('Histogram of Speed Error', fontsize=16)
plt.legend(fontsize=20)
plt.grid(True)
plt.xticks(fontsize=22, rotation=45)
plt.yticks(fontsize=22)
plt.tight_layout()
plt.savefig(plot3_name, format="pdf", bbox_inches="tight")
plt.show()
print(f"Error histogram saved as: {plot3_name}")

# === Plot 4: Loss Over Time ===
plt.figure(figsize=(9, 7))
plt.plot(times, loss, label='Loss', color='purple', lw=2)
plt.xlabel('Time (s)', fontsize=24)
plt.ylabel('Loss', fontsize=24)
#plt.title('Loss Over Time', fontsize=16)
plt.legend(fontsize=16)
plt.grid(True)
plt.xticks(fontsize=22)
plt.yticks(fontsize=22)
plt.tight_layout()
plt.savefig(plot4_name, format="pdf", bbox_inches="tight")
plt.show()
print(f"Loss plot saved as: {plot4_name}")
```
